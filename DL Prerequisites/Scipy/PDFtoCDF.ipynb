```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
x = np.linspace(-6, 6, 1000)
```

```python
fx = norm.pdf(x, loc=0, scale=1)
```

```python
plt.plot(x,fx)
```

```python
Fx = norm.cdf(x, loc=0, scale=1)
```

```python
plt.plot(x, Fx)
```

```python
lofFx = norm.logpdf(x, loc=0, scale=1)
```

```python
plt.plot(x,lofFx)
```
