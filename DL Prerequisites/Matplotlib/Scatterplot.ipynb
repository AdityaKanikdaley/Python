```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.randn(100,2)
plt.scatter(x[:,0], x[:,1]) #-> x[:,0](horizontal) & x[:,1](verticle)
```

```python
x = np.random.randn(200, 2) # by default randn numbers are centered at 0
# to change ( suppose to 3) for 1st 50 elements(rows from index 0 to 50), we write
x[:50] += 3
y = np.zeros(200)
y[:50] = 1
plt.scatter(x[:,0], x[:,1], c=y); # c = color
 
```
